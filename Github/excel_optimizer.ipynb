```python
import pandas as pd
import difflib

# Load your Excel file
file_path = 'common.xlsx'  # Update this with your actual path
df = pd.read_excel(file_path)

# Get the bold names (column headers)
bold_names = df.columns.tolist()

# Optional: create a quick lookup for fixing similar names
def fix_name(name, candidates, cutoff=0.85):
    matches = difflib.get_close_matches(name, candidates, n=1, cutoff=cutoff)
    return matches[0] if matches else name

# Step 1: Normalize all follower names by fixing typos based on bold names
for col in df.columns:
    df[col] = df[col].apply(lambda x: fix_name(str(x).strip(), bold_names) if pd.notna(x) else x)

# Step 2: Build a dictionary of followers under each bold name
result = {name: df[name].dropna().astype(str).tolist() for name in bold_names}

# Step 3: If a follower is also a bold name, append the current bold name under that follower
for name in bold_names:
    for follower in result[name]:
        if follower in bold_names and name not in result[follower]:
            result[follower].append(name)

# Step 4: Convert back to DataFrame
max_len = max(len(v) for v in result.values())
padded_result = {k: v + [None] * (max_len - len(v)) for k, v in result.items()}
final_df = pd.DataFrame(padded_result)

# Step 5: Save to new Excel file
# output_path = 'new_common.xlsx'  # Or any path you want
# final_df.to_excel(output_path, index=False)

print(f"Updated file saved to {output_path}")
final_df.iloc[:,17:]
```
